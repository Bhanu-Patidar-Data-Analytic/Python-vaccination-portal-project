```python
import pandas as pd
from datetime import datetime, timedelta
```

```python
df = pd.read_excel('vaccine_data.xlsx')
df.head()
```

```python
userid = input("Enter your User ID: ")
user = df[df['userid'] == userid]

if user.empty:
    print("User not found")
else:
    name = user.iloc[0]['name']
    first_status = user.iloc[0]['1st dose status']
    first_date = user.iloc[0]['date of 1st dose']
    second_status = user.iloc[0]['2nd dose status']
    second_date = user.iloc[0]['date of 2nd dose']

    print(f'Hi {name},')

    if first_status == 'yes' and second_status == 'yes':
        print('Congratulations! You are fully Vaccinated')
    
    elif first_status == 'no':
        print('Please take your first dose at earliest')
    
    elif first_status == 'yes' and second_status == 'no':
        today = datetime.today()
        first_date = pd.to_datetime(first_date)
        day = (today - first_date).days

        if day > 90:
            print(f'You have taken your first dose on {first_date.strftime('%d/%m/%Y')}, Please take your 2nd dose at earliest')
        else:
            next_dose = first_date + timedelta(days=90)
            print(f'you have taken your first dose on {first_date.strftime('%d/%m/%Y')}, please take your 2nd dose on or after {next_dose.strftime('%d/%m/%Y')}')
```
